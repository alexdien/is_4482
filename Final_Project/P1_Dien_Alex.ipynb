### **1) Is the data mining task a classification or regression task? In a text block explain the difference between the two. **
The data mining task is a classification task because we are trying to predict the category of an input sample based on its features. Our dataset contains a mix of categorical and continuous variables, such as education, age, occupation, etc. and we are predicting them against our target variable, whether the individual earns more or less than $50,000 a year.

The difference between a regression task and a classification task is that a regression task is used to predict a continuous value for an input and primarily focuses more on numbers. For example, if we are given a dataset of the stock market with features such as price, rates, and other continuous variables, we are trying to predict a number outcome. In summary, the main difference is the type of output that is predicted. Classification tasks output a category, while regression tasks output a continuous value.

### Describe the strengths and weaknesses of each metric. 
Strengths of accuracy are: simplicity, easy, intuitive, useful. Weaknesses are: inaccuracy on imbalanced data, biases.

Strengths of precision are: ability to evaluate false positive predictions, correctness of predictions. Weaknesses are: Misleading with imbalanced data, possible false negatives.

Strengths of recall are: ability to identify relevant instances of target, flexibility. Weaknesses are: only takes in false negatives, sensitive to imbalanced data, can be misleading.

Strengths of F1 score: ability to identify true positives and minimize false positives, good for imbalanced data. Weaknesses: does not reveal related information among features.
