```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='r', alpha=0.3)

plt.title("A simple line graph")
plt.xlabel("time")
plt.ylabel("amount")
plt.show()
```

In the graph, I changed the face color to red. I also added labels to the 'x' and 'y' axis with the respective labels 'amount' and 'time'. The alpha parameter changes how dark the shade of the color is.
